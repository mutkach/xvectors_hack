```python
import pandas as pd
```

```python
df=pd.read_csv('data/archive-3/amazon_products.csv')
categories = pd.read_csv('data/archive-3/amazon_categories.csv')
```

```python
final_categories=pd.DataFrame()
```

```python
# TOOLS
```

```python
tool_categories=df[df['title'].str.contains('DEWALT') | df['title'].str.contains('Makita')].category_id.value_counts()[:4]
tool_categories
```

```python
#if tool_categories['id'] not in final_categories['id']:
final_categories=pd.concat([final_categories, categories[categories.id.isin(tool_categories.index)]])
```

### Vacuums and home stuff

```python
vacuum_categories=df[df['title'].str.contains('Dyson')].category_id.value_counts()[:2]
vacuum_categories
```

```python
final_categories=pd.concat([final_categories, categories[categories.id.isin(vacuum_categories.index)]])
```

```python
final_categories
```

```python
# Tent and camping
```

```python
tent_categories = df[df['title'].str.contains('Camping') & df['title'].str.contains('Tent') & (df['price'] > 100)].category_id.value_counts()[:2]
final_categories=pd.concat([final_categories, categories[categories.id.isin(tent_categories.index)]])
```

```python
# Drones
```

```python
drone_categories = df[df['title'].str.contains('Drone') | df['title'].str.contains('DJI') & (df['price'] > 100)].category_id.value_counts()[:1]
final_categories=pd.concat([final_categories, categories[categories.id.isin(drone_categories.index)]])
```

```python
result = df[df['category_id'].isin(final_categories.id) & (df['price'] > 100)]
result = result[~result['title'].str.contains('Kids')]
result.index = range(len(result))

result.to_csv("data/archive-3/amazon_best.csv")
```
